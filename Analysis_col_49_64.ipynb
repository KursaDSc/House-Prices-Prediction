```python
# Replace multiple values in a specific column
replacement_dict = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1}
replacement_dict1 = {'Fin': 3, 'RFn': 2, 'Unf': 1}
feat_set4['KitchenQual'] = feat_set4['KitchenQual'].replace(replacement_dict)
feat_set4['FireplaceQu'] = feat_set4['FireplaceQu'].replace(replacement_dict)
feat_set4['GarageFinish'] = feat_set4['GarageFinish'].replace(replacement_dict1)
feat_set4['GarageQual'] = feat_set4['GarageQual'].replace(replacement_dict)
feat_set4['GarageCond'] = feat_set4['GarageCond'].replace(replacement_dict)
```

```python
feat_set4['FireplaceQu'] = feat_set4['FireplaceQu'].fillna(0).astype('int64')
feat_set4['GarageType'] = feat_set4['GarageType'].fillna('NG')
feat_set4['GarageYrBlt'] = feat_set4['GarageYrBlt'].fillna(0).astype('int64')
feat_set4['GarageFinish'] = feat_set4['GarageFinish'].fillna(0).astype('int64')
feat_set4['GarageQual'] = feat_set4['GarageQual'].fillna(0).astype('int64')
feat_set4['GarageCond'] = feat_set4['GarageCond'].fillna(0).astype('int64')
```

```python
columns = feat_set4.columns.to_list()
columns
```

```python


plt.figure(figsize=(15,75))

for i in range(len(columns)):
    plt.subplot(18,2,i+1)
    plt.hist(feat_set4[columns[i]])
    plt.title(columns[i])

plt.show()
```

```python
plt.figure(figsize=(18,40))

for i, col in enumerate(columns):
    if col != 'SalePrice':
        plt.subplot(8,2,i+1)
        plt.scatter(feat_set4["SalePrice"], feat_set4[col])
        plt.title(f"SalePrice vs {col}")
```

```python
func_df = feat_set4.groupby('Functional')[feat_set4.columns].mean()
func_df = pd.DataFrame(func_df)
func_df.reset_index(inplace= True)
func_df
```

```python
func_df.plot.bar(x='Functional', y='SalePrice', rot=0)

# Add labels and title
plt.xlabel('Functional')
plt.ylabel('SalePrice')
plt.title('Functional Vs. SalePrice')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Functional', y='SalePrice', data=func_df)
plt.xlabel("Functional");
plt.ylabel("SalePrice")
```

```python
grg_df = feat_set4.groupby('GarageType')[feat_set4.columns].mean()
grg_df = pd.DataFrame(grg_df)
grg_df.reset_index(inplace= True)
grg_df
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GarageType', y='SalePrice', data=grg_df)
plt.xlabel("GarageType");
plt.ylabel("SalePrice")
```
